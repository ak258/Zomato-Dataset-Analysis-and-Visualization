## Removing "NEW" , "-" and "/5" from Rate Column


```python
def handlerate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
    
df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

#IN THE RATE COLUMN WE HAVE /5 VALUE,AND ALSO IT HAS STRING VALUES ie NEW,-
created a function handlerate ,so where ever the value is NEW or - THEN IT WILL REPLACE THOSE VALUES WITH NULL VALUE.
AND IF THE NOT THEN THE STRING VALUE WILL SPLIT FROM THE / AND IT WILL BE A LIST ,NOW THE LIST WILL CONTAIN TWO VALUES EX 2.2 AT O INDEX AND /5 AT 1 INDEX
AND NOW WE CREATE VALUE VARIABLE WHICH IS EQUAL TO VALUE OF INDEX 0.

```python
#we can see there are commas in the numbers which restricts the value to be float or int so we will have to remove it.
```

```python
def handleCostfor2(value):
    value=str(value)
    if ',' in value:
        value=value.replace(',','')
        return float(value)
    else :
        return float(value)


df['Costfor2']=df['Costfor2'].apply(handleCostfor2)
df['Costfor2'].unique()
```

### Cleaning rest_type

```python
rest_type=df['rest_type'].value_counts(ascending=False)
rest_type
```

```python
rest_typelessthan1000=rest_type[rest_type<1000]
rest_typelessthan1000
```

```python
def rest_typeothers(value):
    if(value in rest_typelessthan1000):
        return 'others'
    else:
        return value
```

```python
df['rest_type']=df['rest_type'].apply(rest_typeothers)
df['rest_type'].value_counts()
```

```python
df.head()
```

 ### Cleaning Location Column

```python
locationcount=df['location'].value_counts()
locationcount
```

```python
locationcountlessthan300=locationcount[locationcount<300]
locationcountlessthan300
```

```python
def locationasother(value):
    value=str(value)
    if(value in locationcountlessthan300):
        return 'others'
    else:
        return value
```

```python
df['location']=df['location'].apply(locationasother)
df['location'].value_counts()
```

```python
df.head()
```

### Cleaning Cuisines Column

```python
countcuisines=df['cuisines'].value_counts()
countcuisines
```

```python
countcuisineslessthan100=countcuisines[countcuisines<100]
countcuisineslessthan100
```

```python
def cuisinesothers(value):
    if(value in countcuisineslessthan100 ):
        return 'others'
    else:
        return value
    
    
df['cuisines']=df['cuisines'].apply(cuisinesothers)
df['cuisines'].value_counts()
```
